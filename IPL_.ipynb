```python
#1.Finding  matches are being won by no. of runs and no. of wickets. by each team
```

```python
# Filter the data to include only matches that had a result
matches_with_result = match_data.dropna(subset=['result'])
```

```python
# Count the number of matches won by runs and by wickets
result_counts = match_data['result'].value_counts()
```

```python
print("Number of matches won by runs and wickets:")
print(result_counts)
```

```python
#Graph Representation
plt.figure(figsize=(6,4))
result_counts.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Matches Won by Result Type')
plt.xlabel('Result Type')
plt.ylabel('Number of Matches')
plt.xticks(rotation=0)
plt.show()
```

```python
#2.Finding  matches are being won by no. of runs and no. of wickets. in each season
```

```python
# Matches won by runs per season
runs_wins_per_season = match_data[match_data['result'] == 'runs'].groupby('season').size()

# Matches won by wickets per season
wickets_wins_per_season = match_data[match_data['result'] == 'wickets'].groupby('season').size()

# Combine the results into a DataFrame
season_wins = pd.DataFrame({
    'Wins by Runs': runs_wins_per_season,
    'Wins by Wickets': wickets_wins_per_season
}).fillna(0).astype(int)

# Display the combined results
display(season_wins)
```

```python
#Graph Representation
plt.figure(figsize=(10,6))
season_wins.plot(kind='bar', color=['red', 'lightgreen'])
plt.title('Matches Won by run and wicket per season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

```python
#3.Finding and comparing  no. of player of match by virat kohli either batting first or bating second
```

```python
# Filter matches where Virat Kohli was player of the match
kohli_potm = match_data[match_data['player_of_match'] == 'V Kohli']

# Determine if the winning team batted first or second
# A team bats first if the match result is 'runs'
# A team bats second if the match result is 'wickets'
kohli_potm['batting_preference'] = kohli_potm['result'].apply(lambda x: 'Batting First' if x == 'runs' else ('Batting Second' if x == 'wickets' else 'Other'))

# Count the number of Player of the Match awards for each batting preference
kohli_batting_preference_counts = kohli_potm['batting_preference'].value_counts()

# Display the results
print("Virat Kohli's Player of the Match Awards based on Batting Preference:")
display(kohli_batting_preference_counts)

# Graph Representation
plt.figure(figsize=(6, 4))
kohli_batting_preference_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Virat Kohli Player of the Match Awards by Batting Preference')
plt.xlabel('Batting Preference')
plt.ylabel('Number of Awards')
plt.xticks(rotation=0)
plt.show()
```

```python
#5. Creating a pi chart for virat kohli for how many runs by 1 ,2 , 4 and 6
```

```python
# Filter the deliveries data for Virat Kohli
kohli_deliveries = df_deliveries_data[df_deliveries_data['batsman'] == 'V Kohli']

# Filter for runs that are 1, 2, 4, or 6
kohli_runs = kohli_deliveries[kohli_deliveries['batsman_runs'].isin([1, 2, 4, 6])]

# Count the occurrences of each run type
run_counts = kohli_runs['batsman_runs'].value_counts().sort_index()

#Result
print(run_counts)
```

```python
# Pie-Chart Representation
plt.figure(figsize=(8, 8))
plt.pie(run_counts, labels=run_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightcoral', 'lightblue', 'yellowgreen', 'orange'])
plt.title('Distribution of Runs Scored by Virat Kohli (1s, 2s, 4s, 6s)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#2008,2009,10 for each season find out which team choose bat first and field first during winning toss every year
#reset index==> convert series index to dataframe
#rename is used to change column name
match_data["season"]=pd.to_datetime(match_data["date"]).dt.year
season_toss_count_df = match_data.groupby(["season" , "toss_decision"]).size().reset_index().rename(columns={0:"count"})
season_toss_count_df
##converting above series into dataframe
```

```python
#Plotting Graph for Comparision
plt.figure(figsize=(10,6))
ax = sns.barplot(x='season', y = 'count', hue = 'toss_decision',
            data = season_toss_count_df)
for bar in ax.patches:
    height = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,  # horizontal center of the bar
        height / 2,                         # vertical middle of the bar
        f'{int(height)}',                   # value (you can use round(height, 1) for floats)
        ha='center', va='center',           # alignment
        color='white', fontsize=9, fontweight='bold'  # make it visible inside the bar
    )
```
